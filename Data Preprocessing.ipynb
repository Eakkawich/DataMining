```python
t=[]
for i in y:   
    land = metadata[metadata['จังหวัด'].str.contains(i)]
    print(i,len(land))
    if len(land)<= 5000 and len(land)>=3000:
        t.append(i)
    for j in range(8):
        j=j+51
        print('ปีใหม่ '+str(j), len(land[land['ชื่อเทศกาล'].str.contains('ปีใหม่ '+str(j))]))
```
